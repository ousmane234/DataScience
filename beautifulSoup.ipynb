```python
from selenium import webdriver
from bs4 import BeautifulSoup
```

```python
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager 
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
```

```python
driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()))
```

```python
url = "https://www.premierleague.com/stats/top/players/goals"
driver.get(url)
```

```python
print(driver.page_source)
```

```python
soup = BeautifulSoup(driver.page_source ,"html.parser")
headerrow = soup.find("tr")
```

```python
array = []
for a in headerrow.find_all("th"):
	array.append(a.get_text())
print(array)
```

```python
players = []
while(True):
          soup = BeautifulSoup(driver.page_source ,"html.parser")
          page = soup.find_all("tr")[1:] #   ne pas recuperer encore les entetes de la pages
          for i in range(1,len(page)):
                    player = page[i].find_all("td")
                    rang =str(player[0].text).replace("." ,"")
                    nom =str(player[1].text).replace("." ,"").strip()
                    club = str(player[2].text).strip()
                    nationalite = str(player[3].text).strip()
                    nb_but = str(player[4].text).strip()
                    players.append([rang ,nom , club , nationalite, nb_but])
          next_button = WebDriverWait(driver ,10).until(EC.element_to_be_clickable((By.CSS_SELECTOR,".paginationNextContainer"))) 
          # print(next_button.get_attribute("class"))   
          if "inactive" in next_button.get_attribute("class") :
                    break 
          next_button.click()       
```

```python
del(array[-1])
array
```

```python

import pandas as pd
df = pd.DataFrame(players , columns = array )
df
```

```python
#### recuperation des meillerus  buteurs :####
# on change le type de Rank en int  pour faire le trie en fonction du  rang
df["Rank"] = df["Rank"].apply(lambda x : int(x))
# trier les joueurs de par leur rang !
df_sorted = df.sort_values("Rank"  , ascending = True )
# recuperer les 50 meilleurs buteurs 
df_sorted.head(50)


```

```python
####  Recuperer les données au fromat excel
df_excel = df_sorted.to_excel("./buteur.xlsx" , sheet_name = "Buteurs premier League 2021-2022")
```
